```python
import numpy as np #常见计算用库，适合大批量处理数据尤其是数组格式的数据
import matplotlib.pyplot as plt #python里的ggplothhh
```

```python
x_train=np.array([1.0,2.0]) 
y_train=np.array([300.0,500.0]) #创建两个数组
print(f"x_train={x_train}")
print(f"y_train={y_train}") #f""格式化字符串，可以在字符串中直接嵌入变量和表达式
```

```python
print(f"x_train.shape:{x_train.shape}")
m=x_train.shape[0] #shape,一般是查看数组的形状，0是数组在第一个唯独上的大小，因为这里是一维数组所以直接显示的是元素的数目
print(f"Number of train examples is:{m}")
```

```python
plt.scatter(x_train,y_train,marker="x",c="y") #scatter散点图，横坐标，纵坐标，用什么符号编辑，颜色
plt.title("Houseing price") 
plt.ylabel("Price (in 1000s of dollars)")
plt.xlabel("Size (1000 sqft)")
plt.show() #显示对应的图表
```

```python
w = 100 #line model wight 也就是斜率
b = 100 #bias 偏置也就是结局
#这里是一个非常非常标准的函数的写法，可以学习一下
def compute_model_output(x,w,b):
    """
    function:computes the prediction of a line model
    Args: #参数
        x(ndarry(m,)):Data,m examples
        w,b (scalar) :model paramters
    returns #返回值
        y (ndarray (m,)):target values
    """
    m=x.shape[0] #获取数据点的数量
    f_wb =np.zeros(m) #初始化数据，创建一个全0的数组，用来后续存放预测值
    for i in range(m):
        f_wb[i]=w*x[i]+b #预测值
    return f_wb
```

```python
tmp_f_wb=compute_model_output(x_train,w,b)
plt.plot(x_train,tmp_f_wb,c="b",label="our prediction")
plt.scatter(x_train,y_train,marker="x",c="y",label="Actual Value")
plt.title("Houseing Prices")
plt.ylabel("Price (in 1000s of dollars)")
plt.xlabel("Size (1000 sqrt)")
plt.legend()
plt.show()
```
